**Попробуем сгенерить новый фич как расстояние от центра. Выберем координату центра**

```python
center = {'lat': 47.62063, 'long': -122.22001}
```

```python
earth_radius = 6371
```

```python
def get_dist(lat, long, center_dict, earth_radius):
    cos_d = np.sin(np.radians(lat)) * np.sin(np.radians(center_dict['lat'])) + np.cos(np.radians(lat)) * np.cos(np.radians(center_dict['lat'])) * np.cos(np.radians(long - center_dict['long']))
    return np.arcsin(cos_d) * earth_radius
```

```python
distances = get_dist(house_df['lat'], house_df['long'], center, earth_radius)
```

```python
distances = np.zeros((house_df.shape[0]))
for i in range(house_df.shape[0]):
    distances[i] = geopy.distance.geodesic((center['lat'], center['long']), (house_df['lat'][i], house_df['long'][i])).km
```

```python
dists = pd.Series(data=distances, name='dists_to_center')
```

```python
house_df = house_df.merge(dists, left_index=True, right_index=True)
```

# **зависимость центы от дистанции до центра**

```python
fig = px.scatter(house_df, x='dists_to_center', y='price')
fig.write_html('dist_to_center_VS_price.html')
```
